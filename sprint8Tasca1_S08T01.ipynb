```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

Dataset 

Source: https://gitlab.com/djh_or/2018-world-cup-stats/blob/master/world_cup_2018_stats.csv

```python
worldcup = pd.read_csv('world_cup_2018_stats.csv')
pd.set_option('display.max_columns', None)
worldcup
```

```python
worldcup.columns
```

```python
# Null hyphotesis: Ho:Distance Covered Mean = 103 (La distancia media percurrida por los jugadores es 103 km)
# Alternate Hypothesis: H1: Distance Covered Mean != 103 
# Alpha = 0.05 or 5%

from scipy.stats import ttest_1samp

distance_mean = np.mean(worldcup['Distance Covered km'])
print("Mean Distance:")
print(distance_mean)

tset, pvalue = ttest_1samp(worldcup['Distance Covered km'], 103)
print("p-value:")
print(pvalue)
print("")
if pvalue < 0.05:  
    print("Null hypothesis rejected.")
else:
    print("Null hypothesis accepted.")
```

```python
from scipy import stats as st
from bioinfokit.analys import get_data

st.ttest_1samp(worldcup['Distance Covered km'], popmean=103) # T statistic and p-value
```

```python
# One sample t-test using bioinfokit
from bioinfokit.analys import stat

res = stat()
res.ttest(worldcup, test_type=1, res='Distance Covered km', mu=5)
print(res.summary)
```

```python
# Analysis of Variance Test (ANOVA)
# Tests whether the means of two or more independent samples are significantly different.
# En este ejercício 

# Null hyphotesis: Ho: the means of the samples are equal.
# Alternate Hypothesis: H1: one or more of the means of the samples are unequal.
```

```python
worldcup.columns
```

```python
# ANOVA - Test 1
from scipy.stats import f_oneway

data1 = worldcup['Goals For']
data2 = worldcup['Ball possession %']
data3 = worldcup['Fouls Committed']
stat, p = f_oneway(data1, data2, data3)

print('stat=%.6f, p=%.6f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

```python
#Test 2
data1 = worldcup['Tackles']
data2 = worldcup['Blocks']
data3 = worldcup['Clearances']
stat, p = f_oneway(data1, data2, data3)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

```python
#Test 3
data1 = worldcup['Passes']
data2 = worldcup['Passes Completed']
data3 = worldcup['Pass Accuracy %']
stat, p = f_oneway(data1, data2, data3)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```

```python
#Test 4
data1 = worldcup['Attempts']
data2 = worldcup['On-Target']
data3 = worldcup['Off-Target']
stat, p = f_oneway(data1, data2, data3)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```
